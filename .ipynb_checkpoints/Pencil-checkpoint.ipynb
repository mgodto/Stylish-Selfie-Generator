```python
import cv2
import matplotlib.pyplot as plt
import imageio


filename = 'C:/Users/gaoji/OneDrive/桌面/000009.png'
photo = imageio.imread(filename)
print("in function:" + filename)
gray_image = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY) 
gray_image = np.array(gray_image)

f, axs = plt.subplots(1,5,figsize=(30,30))

size = 15

plt.subplot(1,5,1)
plt.title('gray scaled image', fontsize=size)
plt.xticks([])
plt.yticks([])
plt.imshow(gray_image,cmap='gray')

plt.subplot(1,5,2)    
plt.title('inverted gray scaled image', fontsize=size)
plt.xticks([])
plt.yticks([])
inverted_gray_image = 255 - gray_image
plt.imshow(inverted_gray_image,cmap='gray')


plt.subplot(1,5,3) 
plt.title('blurred inverted gray scaled image', fontsize=size)
plt.xticks([])
plt.yticks([])
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21),0) 
plt.imshow(blurred_img,cmap='gray')


plt.subplot(1,5,4) 
plt.title('inverted blurred inverted gray scaled image', fontsize=size)
plt.xticks([])
plt.yticks([])
inverted_blurred_img = 255 - blurred_img
plt.imshow(inverted_blurred_img,cmap='gray')


plt.subplot(1,5,5) 
plt.title('Final result', fontsize=size)
pencil_sketch_IMG = cv2.divide(gray_image, inverted_blurred_img, scale = 256.0)
plt.xticks([])
plt.yticks([])
plt.imshow(pencil_sketch_IMG,cmap='gray')
```
